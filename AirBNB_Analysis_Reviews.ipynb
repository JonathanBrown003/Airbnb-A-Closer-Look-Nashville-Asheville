```python
# Imports 

import pandas as pd 
import numpy as np
```

```python
# Reading the Nashville data 
nash_data = pd.read_csv("listings.csv")
nash_data.head()
```

```python
#  Nashville data shape
nash_data.shape
```

```python
# Reading the Asheville data 
ashe_data = pd.read_csv("Asheville_listings.csv")
ashe_data.head()
```

```python
# Asheville data shape
ashe_data.shape
```

```python
# Checking for duplicates 
nash_unique = nash_data.drop_duplicates(subset=['id'])
nash_unique.shape 
# all unique entries 

ashe_unique = ashe_data.drop_duplicates(subset=['id'])
ashe_unique.shape
# all unique entries 
```

```python
nash_unique
```

```python
# Looking at the columns to drop 
nash_unique.columns

# get ridding of scrape_id, last_scraped, host_has_profile_pic, picture_url, host_thumbnail_url, host_picture_url, host_ neighborhood, bathrooms, neighbourhood_group_cleansed, calendar_updated, minimum_minimum_nights, maximum_minimum_nights, 
# minimum_maximum_nights, maximum_maximum_nights, minimum_nights_avg_ntm  , maximum_nights_avg_ntm, has_availability, "calendar_last_scraped", "license", "month", 
nash_columns = nash_unique.drop(["scrape_id", "last_scraped", "host_has_profile_pic", "picture_url", "host_thumbnail_url", "host_picture_url", "host_neighbourhood", "bathrooms",\
     "neighbourhood_group_cleansed", "calendar_updated", "minimum_minimum_nights", "maximum_minimum_nights", "minimum_maximum_nights", "maximum_maximum_nights",\
    "minimum_nights_avg_ntm", "maximum_nights_avg_ntm", "has_availability", "availability_30", "availability_60", "availability_90", "availability_365", "calendar_last_scraped", "license"], axis=1)
```

```python
# Dropping columns for Asheville dataset 
ashe_columns = ashe_unique.drop(["scrape_id", "last_scraped", "host_has_profile_pic", "picture_url", "host_thumbnail_url", "host_picture_url", "host_neighbourhood", "bathrooms",\
     "neighbourhood_group_cleansed", "calendar_updated", "minimum_minimum_nights", "maximum_minimum_nights", "minimum_maximum_nights", "maximum_maximum_nights",\
    "minimum_nights_avg_ntm", "maximum_nights_avg_ntm", "has_availability", "availability_30", "availability_60", "availability_90", "availability_365", "calendar_last_scraped", "license"], axis=1)
```

```python
# dropping the null rows 
nash_rows = nash_columns.dropna()
nash_rows.head()
```

```python
# Asheville columns 
ashe_rows = ashe_columns.dropna()
ashe_rows.shape
```
